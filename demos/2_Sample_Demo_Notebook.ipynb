```python
import numpy as np
import pandas as pd
from scipy import spatial
from two_sample import Two_Sample
np.set_printoptions(threshold=np.nan)
```

```python
#read in the data
print('This demo notebook will demonstrate two sample testing. First, read in the data about the salaries of men vs those of women.')
x=pd.read_csv('salary_csv.csv')
men=[]
women=[]
```

```python
#create data matrices for men's and women's salaries
print('Next, store the data in lists.')
for i in range (len(x)):
    if x.loc[i][0]=='M':
        men.append(x.loc[i][1])
    elif x.loc[i][0]=='F':
        women.append(x.loc[i][1])
```

```python
print('Convert lists to numpy arrays.')
men=np.asarray(men)
women=np.asarray(women)
```

```python
print('The list for salaries of men looks like: \n',men)
```

```python
print('The list for salaries of women looks like: \n',women)
```

```python
print('Create an instance of the Two_Sample object, specifying which test you would like to feed the transformed matrices thorugh.')
print('The default test is unbiased dcorr. For this example, use mgc.')
two=Two_Sample(ind_test='mgc')
```

```python
print('The transform function takes the two data matrices and outputs a concatenated data matrix (A) and a label matrix (B).')
print('The label matrix marks all data points from men with a 0 and all data points from women with a 1.')
a,b=two.transform(men,women)
```

```python
print('The concatenated data matrix A looks like: \n',a)
```

```python
print('The label matrix B looks like: \n',b)
```

```python
t_stat,p_val=two.independence(men,women)
```

```python
print('The test statistic is',t_stat[0])
```

```python
print('The p value is', p_val[0])
```
